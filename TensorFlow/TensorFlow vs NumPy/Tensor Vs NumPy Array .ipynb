## Use to NumPy

```python
shape = 5000
```

```python
# Create NumPy Array

# Create a 5000x5000 NumPy array with uniform distribution between 0 and 1
np_array = np.random.uniform(low=0.0, high=1.0, size=(5000, 5000))
```

```python
start = time.time()

for i in range(10):  # Loop ten times
    np.matmul(np_array, np_array)  # Matrix multiplication

end = time.time()

time_period = end - start
print(f'{time_period} seconds')
```

* NumPy not use GPU

## Use to Tensor

```python
# Create Tensor Array

# Create a 5000x5000 TensorFlow tensor with uniform distribution between 0 and 1
tf_array = tf.random.uniform(shape=(5000, 5000), minval=0.0, maxval=1.0)
```

```python
start = time.time()

for i in range(10):  # Loop ten times
    tf.matmul(tf_array, tf_array)  # Matrix multiplication

end = time.time()

time_period = end - start
print(f'{time_period} seconds')
```

* Tensor use GPU
