```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential([ # Use Sequential API
    tf.keras.layers.Conv2D(10, (5,5), activation='relu', input_shape=(64, 64, 3) ), # Create 1st Convolutional Layer (filters=10, filter/Kernal size=(5,5, stride defult(1,1))
    tf.keras.layers.MaxPooling2D((3,3), strides=(3,3)), # Create 1st Pooling Layer (pooling size = (3,3))

    tf.keras.layers.Conv2D(16, (5,5), activation='relu'),  # Create 2st Convolutional Layer
    tf.keras.layers.MaxPool2D((2,2), strides=(2,2)), # Create 2st Pooling Layer

    # Do Flattern (Convert one array)
    tf.keras.layers.Flatten(),

    # Using Dense Layer
    tf.keras.layers.Dense(128, activation='relu'), # 1st Dense Layer --> Num of neouron = 128
    tf.keras.layers.Dense(16, activation='relu'), # 1st Dense Layer --> Num of neouron = 16
    tf.keras.layers.Dense (1, activation='sigmoid') # Output layer (Assum Binary classification problem becouse output is one neouron )

])
```

```python
# Model Summary
model.summary()
```
