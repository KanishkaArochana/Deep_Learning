```python
from tensorflow.keras.utils import img_to_array, load_img
# img_to_array ---> After load img convert numarical array
# load_img --> Load image
```

```python
# Image Load
img = load_img('/content/elephant.jpg')
```

```python
# After load Show img
img.show()
```

```python
# Convert Numarical Array and assing x
x = img_to_array(img)
```

```python
x
```

```python
# x shape
x.shape
```

```python
# x Reshape (rank = 4)
x = x.reshape((1,) + x.shape)
```

**Apply Data Augmentation**

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# ImageDataGenerator -->
```

```python
# Create Image Data Generate Object
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
)
```

```python
# Apply Image to Data Augmentation and Save Img
i = 0
for batch in datagen.flow(x, batch_size=1, save_to_dir='/content/augmented_images', save_prefix='elephant', save_format='jpeg'):
  i += 1
  if i > 10: # 10 Augment image
    break
```
